### Token Sort Ratio

```python
# calculates the ratio by matching the alphabets in a order

strA = 'Neil Nitin Mukesh'
strB = 'Nitin Neil Mukesh'
tsratio = fuzz.token_sort_ratio(strA,strB)
print('Token Sort Ratio is',tsratio)
ratio = fuzz.ratio(strA,strB)
print('Ratio is',ratio)
```

Fuzzy Wuzzy library provide first API which is very useful in doing string matching for applications like NLP, it provides us with variety of string matching functions that can be used in different scenarios.
1. <b>Ratio Function</b> finds out the similarity ratio between the two string using the Levenshtein distance formula.
2. <b>Partial Ratio Function</b> is used to perform sub-string matching. This function is typically useful while matching people's name
3. <b>Token Sort Ratio</b> function sorts the string alphabetically and then do matching on the alphabetically sorted versions of the string.

### Token Set Ratio

```python
a = 'truth or dare'
b = 'truth or or dare'
tsratio = fuzz.token_sort_ratio(a,b)
print('Token Sort Ratio is',tsratio)

tsetratio = fuzz.token_set_ratio(a,b)
print('Token Set Ratio is',tsetratio)
```

```python
a = 'Neil Nitin'
b = 'Neil Nitin Nitin'
tsratio = fuzz.token_sort_ratio(a,b)
print('Token Sort Ratio is',tsratio)

tsetratio = fuzz.token_set_ratio(a,b)
print('Token Set Ratio is',tsetratio)
```

### WRatio
WRatio is a string matching function that has the ability to ignore the uppercase, lowercase and also alpha-numeric characters.

```python
a = 'I Love Myself'
b = 'I am loving myself'
wratio = fuzz.WRatio(a,b)
print('W Ratio is',wratio)
```
