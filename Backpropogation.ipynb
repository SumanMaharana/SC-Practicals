```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
data = load_iris()
```

```python
x = data.data
x
```

```python
y = data.target
y
```

```python
res = np.zeros((y.size,3), dtype = int)
res[np.arange(y.size),y] = 1
res
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, res, test_size = 20, random_state = 1)
```

```python
y_train.size
```

```python
learning_rate = 0.1
iteration = 5000
n = y_train.size
input_size = 4    #input layer
hidden_size = 2   #hidden layer
output_size = 3   #output layer
results = pd.DataFrame(columns=['mse'])
```

```python
results
```

```python
np.random.seed(10)
w1 = np.random.normal(scale = 0.5, size = (input_size, hidden_size))
```

```python
w1
```

```python
w2 = np.random.normal(scale = 0.5, size = (hidden_size, output_size))
w2
```

```python
def sigmoid(x):
    return 1/(1+np.exp(-x))
```

```python
def mean_squared_error(ypred, ytrue):
    return ((ypred - ytrue)**2).sum()/(2*ypred.size)
```

```python
for itr in range(iteration):
    #Forward Propogation on hidden layer
    z1 = np.dot(x, w1)  #dot product
    a1 = sigmoid(z1)

    #Forward Propogation on output layer
    z2 = np.dot(a1, w2)
    a2 = sigmoid(z2)

    #Calculating the error
    mse = mean_squared_error(a2, res)
    results = results.append({'mse':mse}, ignore_index = True)
    
    #Back Propogation Phase
    e1 = a2 - res #Calculation of error
    dw1 = e1 * a2 * (1-a2) #delta between hidden and output

    e2 = np.dot(dw1, w2.T)
    dw2 = e2 * a1 * (1-a1) #delta between input and hidden
    
    #Updating the weights
    w2_update = np.dot(a1.T, dw1)/n
    w1_update = np.dot(x.T, dw2)/n

    w2 = w2 - learning_rate * w2_update
    w1 = w1 - learning_rate * w1_update
```
